```python
import pandas as pd
```

```python
# Load your dataset
df = pd.read_csv("Data/raw/IT Troubleshouting DATAset - Feuille 1.csv")
df
```

```python
df.rename(columns={
    df.columns[0]: "Customer_Issue",
    df.columns[1]: "Tech_Response"
}, inplace=True)
df
```

```python
# 1. Drop duplicate issues
df = df.drop_duplicates(subset="Customer_Issue", keep="first")
df
```

```python
# Remove leading/trailing spaces in column names and values
df.columns = df.columns.str.strip()
df["Customer_Issue"] = df["Customer_Issue"].astype(str).str.strip()
df["Tech_Response"] = df["Tech_Response"].astype(str).str.strip()
```

```python
# Drop rows where either column is missing
df.dropna(subset=["Customer_Issue", "Tech_Response"], inplace=True)
```
